# Continuous uniform distribution (連続一様分布)

```python
import pymc as pm
import arviz as az
```

```python
with pm.Model() as model:
    d  = pm.Uniform('Continuous Uniform Distribution', lower=1, upper=2)
    trace  = pm.sample(4000, chains=4)
```

```python
az.summary(trace)
```

```python
az.plot_trace(trace, compact=True)
```

# Discrete uniform distribution (離散一様分布)

```python
with pm.Model() as model:
    d  = pm.DiscreteUniform('Discrete Uniform Distribution', lower=1, upper=6)
    trace  = pm.sample(4000, chains=4)
```

```python
az.summary(trace)
```

```python
az.plot_trace(trace, compact=True)
```
